```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('car_purchasing.csv',encoding='latin-1')
```

```python
data.head()
```

```python
df = data['customer name'].str.split(' ', expand=True)
df.columns = ['name1', 'name2', 'name3']
data = pd.concat([data, df], axis=1)
```

```python
df = data['customer e-mail'].str.split('.', expand=True)
df.columns = ['email1', 'email2', 'email3','email4','email5']
df = pd.concat([df, df['email1'].str.split('@',expand=True)], axis=1)
df = pd.concat([df, df['email2'].str.split('@',expand=True)], axis=1)
df = pd.concat([df, df['email3'].str.split('@',expand=True)], axis=1)
df = pd.concat([df, df['email4'].str.split('@',expand=True)], axis=1)
df = pd.concat([df, df['email5'].str.split('@',expand=True)], axis=1)
names = ['email' + str(i) for i in range(len(df.columns))]
df.columns = names
data = pd.concat([data, df], axis=1)
```

```python
for column in data.columns:
    if data[column].dtypes == 'object':
        try:
            data[column] = data[column].astype(float)
        except:
            data[column] = (pd.factorize(data[column])[0])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
x = data.drop('car purchase amount',axis=1)
y = data['car purchase amount']
```

```python
from sklearn.ensemble import RandomForestRegressor
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.30, random_state=42)
model = RandomForestRegressor()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```
